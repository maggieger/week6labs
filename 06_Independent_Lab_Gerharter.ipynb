### Set up lists/variables

```python
first_names = ["Carmen",
               "Sydney",
               "Neil",
               "Marcus"]
last_names = ["Berzatto",
              "Adamu",
              "Fak",
              "Brooks"]
majors = ["Anthropology",
         "History",
         "Biology",
         "Forestry"]
gpas = [3.7,
        3.5,
        3.0,
        3.9]
academic_violations = ["copying homework",
                       "cheating on exam",
                       "plagiarism",
                       "cheating on quiz",
                       "writing paper for another student"]
student_violations = ["none",
                      "none",
                      "none",
                      "none"]

student_added = False  #keeps track of whether a student has been added
```

# Create Nested Loop

LOOP WHILE TRUE  

Create a menu system  
Implement functions for:  
 - Querying student info  
 - adding new students  
 - calculating GPA metrics  
 - assigning academic violations  
 - quit program

```python
print("\n Welcome to the Student Management System")

while True:
    print("\n Menu")
    print("1. Student Information Query")
    print("2. Student Information Entry")
    print("3. Query GPA Metrics")
    print("4. Academic Violations")
    print("5. Quit Program")

    selection = input("Select your option")

    if selection == "1":
        print("\n Student list:")
        for i in range(len(first_names)):
            print(f"{i+1}. {first_names[i]} {last_names[i]} | Major: {majors[i]} | GPA: {gpas[i]} | Violation: {student_violations[i]}")
        
    elif selection == "2":
        print("\n Add new student information")
        first_name = input("First name:")
        first_names.append(first_name)
        student_added = True
        last_name = input("Last name:")
        last_names.append(last_name)
        major = input("Major:")
        majors.append(major)
        gpa = input("GPA:")
        gpas.append(gpa)
        student_violations.append("None")

    elif selection == "3":
        print("\n Query GPA Metrics")
        gpas = [float(gpa) for gpa in gpas]
        max_gpa = max(gpas)
        print(f"Highest GPA: {max_gpa}")
        min_gpa = min(gpas)
        print(f"Lowest GPA: {min_gpa}")
        avg_gpa = (sum(gpas)/len(gpas))
        print(f"Average GPA: {avg_gpa}")

    elif selection == "4":
        print("\n Student list:")
        for i in range(len(first_names)):
            print(f"{i+1}. {first_names[i]} {last_names[i]}")

        student_index = int(input("Select a student by number:"))-1
        if 0 <= student_index <len(first_names):
            print("\n Select the violation:")
            for i, v in enumerate(academic_violations):
                print(f"{i+1}. {v}")

            violation_index = int(input("Enter academic violation by number")) -1
            if 0 <= violation_index < len(academic_violations):
                student_violations[student_index] = academic_violations[violation_index]
                print(f"{first_names[student_index]} {last_names[student_index]} has been marked for: {academic_violations[violation_index]}")
            else:
                print("Error. There is no violation by this number.")

        else:
            print("Invalid student number.")


    elif selection == "5":
        print("\n Program is shut down. Goodbye.")
        break

    else:
        print("\n Invalid selection. Please try again.")
```
